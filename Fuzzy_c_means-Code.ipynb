```python
#Initialising fuzzy c_means variables

#Length of df
N=len(df)

#No of features
n=2

#No of clusters
c =5

#Fuzzy parameter
m=2        

#Error threshold
e=0.001

L=0

Z=df.values
Z=np.array(Z)
Z=Z.transpose()
print(Z)
```

```python
#Membership function
#Randomly generating a list of length "c"
#Then normalizing the list
#Then appending the normalized list in an empty array U
U=np.empty((c,1),dtype=int)
for k in range (N):
    random_num_list = [random.random() for i in range(c)] 
    summation = sum(random_num_list)
    temp_list = [x/summation for x in random_num_list]
    temp_list=np.array(temp_list)
    temp_list = np.reshape(temp_list,(c,1))
    global U
    U=np.append(U,temp_list,axis=1)
#Since first column is zeros, we remove that
U=U[:,1:]
print(U)
```

```python
#Initializing an empty array U_updated of c rows and 800 columns.
U_updated=np.zeros(((c,800)),dtype=float)
print(U_updated)
```

```python
while True: 
    global U
    global U_updated
    global A
    global B
    global J
    diff=0
#Cluster centroids V[i]
    print("Initial matrix:", U)
    U_updated=np.zeros(((c,800)),dtype=float)
    V=[]
    def clustermeans(U,Z):
        global numer
        global denom
        global V
        for i in range (c):
            numer=0
            denom=0
            for k in range (N):
                numer+= math.pow(U[i][k],m)*Z[:,k].reshape(2,1)     
                denom+=  math.pow(U[i][k],m) 
                                       
            q = (numer)/(denom)
 
            
                                                                    
            V.append(q)  
        V=(np.array(V).transpose()).reshape(2,c)
                                  
        return V
    print("The cluster centroids are:",clustermeans(U,Z))
    

#Distance
    d=[]
    def D(i,k):
        h =np.subtract((Z[:,k].reshape(2,1)),(V[:,i]).reshape(2,1))
        d= (h.transpose()).dot(h)                      
        return d     
        
    print("The distance is:",D(1,1))
    
    J=0
    for i in range (1,c):
        for k in range (1,N):
            J += (math.pow(U[i][k],m))*(abs(D(i,k)))
    print("The cost function value is:",J,  c)
    

#Updating matrix
    k_list=[]
    for i in range(c):
        for k in range(N):
            if (D(i,k)>0):
                A=0
                B=0
                for j in range(c):
                    A=A+math.pow((D(i,k)/D(j,k)),(1/(m-1)))
                U_updated[i][k]=1/A

            else:
                k_list.append(k)
                current_r=i
                for row in range(c):
                    if (row == current_r):
                        U_updated[row][k]=1
                    else:
                        U_updated[row][k]=0


    
    print("New matrix", U_updated)  

    
    
    L=L+1
    print(L)
    diff=np.subtract(U, U_updated)
    print("The difference is:",linalg.norm(diff),diff)
    U=U_updated
    if (linalg.norm(diff)<e):
        break
    
```
